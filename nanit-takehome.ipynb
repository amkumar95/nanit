```python
import pandas as pd
import json
 
# Load JSON data into a variable
file_path = 'shippingdata.json'
with open(file_path, 'r') as file:
    data = json.load(file)
 
# Normalize the main part of the JSON data
orders_df = pd.json_normalize(data['Order'], 
                              record_path=['Dispatches'], 
                              meta=['OrderId', 'OrderNumber', 'OrderDate', 'OrderSource', 'CurrencyCode', 'Total', 'TotalTax', ['ShippingAddress', 'AddressLine1'], ['ShippingAddress', 'AddressLine2'], ['ShippingAddress', 'Town'], ['ShippingAddress', 'Postcode'], ['ShippingAddress', 'CountryCode']],
                              record_prefix='Dispatches_',
                              max_level=3)
 
# Function to explode and flatten nested JSON columns
def explode_and_flatten(df, column):
    df = df.explode(column)
    flattened_col_df = pd.json_normalize(df[column])
    flattened_col_df.columns = [f"{column}_{subcolumn}" for subcolumn in flattened_col_df.columns]
    df = df.drop(columns=[column]).reset_index(drop=True)
    df = pd.concat([df, flattened_col_df.reset_index(drop=True)], axis=1)
    return df
 
# To extract Nested Columns
nested_columns = ['Dispatches_DispatchedLines']
 
for column in nested_columns:
    if column in orders_df.columns:
        orders_df = explode_and_flatten(orders_df, column)
 
# Convert OrderDate to datetime format
orders_df['OrderDate'] = pd.to_datetime(orders_df['OrderDate'])
 
# Renaming columns for better readability
orders_df.rename(columns={
    'ShippingAddress.AddressLine1': 'AddressLine1',
    'ShippingAddress.AddressLine2': 'AddressLine2',
    'ShippingAddress.Town': 'Town',
    'ShippingAddress.Postcode': 'Postcode',
    'ShippingAddress.CountryCode': 'CountryCode'
}, inplace=True)
 
# Display the resulting DataFrame
orders_dispatch = orders_df
orders_dispatch.head()
```

```python
import mysql.connector
from mysql.connector import Error
import json
import pandas as pd
import sqlalchemy

# Define MySQL database connection details
db_user = 'sql5709445'
db_password = 'MfzQuM33xk'
db_host = 'sql5.freesqldatabase.com'
db_name = 'sql5709445'

try:
    # Establish a connection to the MySQL database
    connection = mysql.connector.connect(
        host=db_host,
        user=db_user,
        password=db_password,
        database=db_name
    )

    if connection.is_connected():
        db_Info = connection.get_server_info()
        print("Connected to MySQL Server version ", db_Info)
        cursor = connection.cursor()

        # Create orders table
        create_orders_table = """
        CREATE TABLE IF NOT EXISTS orders (
            OrderDate DATETIME,
            OrderId INT PRIMARY KEY,
            OrderNumber VARCHAR(255),
            OrderSource VARCHAR(255),
            Total DECIMAL(10, 2),
            TotalTax DECIMAL(10, 2),
            CurrencyCode VARCHAR(10)
        )
        """
        cursor.execute(create_orders_table)

        # Create shipping details table
        create_shipping_details_table = """
        CREATE TABLE IF NOT EXISTS shipping_details (
            OrderId INT,
            AddressLine1 VARCHAR(255),
            AddressLine2 VARCHAR(255),
            CountryCode VARCHAR(10),
            FirstName VARCHAR(100),
            LastName VARCHAR(100),
            Postcode VARCHAR(20),
            Town VARCHAR(100),
            FOREIGN KEY (OrderId) REFERENCES orders(OrderId)
        )
        """
        cursor.execute(create_shipping_details_table)

        # Create dispatch table
        create_dispatch_table = """
        CREATE TABLE IF NOT EXISTS dispatch (
            OrderId INT,
            Carrier VARCHAR(255),
            Date DATE,
            Reference VARCHAR(255),
            TrackingNumber VARCHAR(100),
            TrackingURL VARCHAR(255),
            ProductCode VARCHAR(255),
            ProductDescription TEXT,
            Quantity INT,
            SerialNumbers TEXT,
            FOREIGN KEY (OrderId) REFERENCES orders(OrderId)
        )
        """
        cursor.execute(create_dispatch_table)

        # Create orderlines table
        create_orderlines_table = """
        CREATE TABLE IF NOT EXISTS orderlines (
            OrderId INT,
            ProductCode VARCHAR(255),
            ProductDescription TEXT,
            Quantity INT,
            UnitCost DECIMAL(10, 2),
            FOREIGN KEY (OrderId) REFERENCES orders(OrderId)
        )
        """
        cursor.execute(create_orderlines_table)

        # Commit the transaction
        connection.commit()

except Error as e:
    print("Error while connecting to MySQL:", e)

finally:
    # Closing database connection
    if 'connection' in locals() and connection.is_connected():
        cursor.close()
        connection.close()
        print("MySQL connection is closed")
```

```python
import mysql.connector
from mysql.connector import Error
import pandas as pd
from sqlalchemy import create_engine

# Define MySQL database connection details
db_user = 'sql5709445'
db_password = 'MfzQuM33xk'
db_host = 'sql5.freesqldatabase.com'
db_name = 'sql5709445'

# Function to convert list columns to strings
def convert_lists_to_strings(df):
    for column in df.columns:
        if df[column].apply(lambda x: isinstance(x, list)).any():
            df[column] = df[column].apply(lambda x: ','.join(map(str, x)) if isinstance(x, list) else x)
    return df

try:
    # Establish a connection to the MySQL database using SQLAlchemy engine
    engine = create_engine(f"mysql+mysqlconnector://{db_user}:{db_password}@{db_host}/{db_name}")

    df_dispatched = convert_lists_to_strings(df_dispatched)
    
    # Assuming df_orders, df_shipping, df_dispatched, df_orderlines are defined DataFrames

    # Load data into orders table
    df_orders.to_sql(name='orders', con=engine, if_exists='replace', index=False)

    # Load data into shipping details table
    df_shipping.to_sql(name='shipping_details', con=engine, if_exists='replace', index=False)

    # Load data into dispatch table
    df_dispatched.to_sql(name='dispatch', con=engine, if_exists='replace', index=False)

    # Load data into orderlines table
    df_orderlines.to_sql(name='orderlines', con=engine, if_exists='replace', index=False)

    print("Data loaded successfully.")

except Error as e:
    print("Error while connecting to MySQL:", e)

finally:
    if 'engine' in locals():
        engine.dispose()  # Dispose of the engine to close all connections

    print("MySQL connection is closed")
```
